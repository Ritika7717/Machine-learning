```python
# step 1: import pandas Library
import pandas as pd
```

```python
# step 2: Read dataset
data=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Admission%20Chance.csv')
```

```python
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# step 3: define problems
data.columns
```

```python
Y=data['Chance of Admit ']
```

```python
X=data[['Serial No', 'GRE Score', 'TOEFL Score', 'University Rating', ' SOP',
       'LOR ', 'CGPA', 'Research',]]
```

```python
data.shape,X.shape,Y.shape
```

```python
# step 4:Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y)
data.shape,X_train.shape,X_test.shape
```

```python
# step:5 Select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# step 6: Train
model.fit(X_train,Y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# step 6: predict
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

```python
# step 6:Evaluation
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(Y_test,Y_pred)
```
